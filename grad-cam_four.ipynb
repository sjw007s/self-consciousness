```python
class LRCN(nn.Module):
    def __init__(self):
        super(LRCN, self).__init__()
        self.layer_1 = nn.Conv2d(3,5,5,stride=2)
        self.layer_2 = nn.Conv2d(3,5,5,stride=2)
        self.layer_3 = nn.Conv2d(3,5,5,stride=2)
        self.layer_4 = nn.Conv2d(3,5,5,stride=2)
        self.layer_5 = nn.Conv2d(3,5,5,stride=2)
        self.layer_6 = nn.Conv2d(3,5,5,stride=2)
        self.layer_7 = nn.Conv2d(3,5,5,stride=2)
        self.layer_8 = nn.Conv2d(3,5,5,stride=2)
        self.layer_9 = nn.Conv2d(3,5,5,stride=2)
        self.layer_10 = nn.Conv2d(3,5,5,stride=2)
        self.layer_11 = nn.Conv2d(3,5,5,stride=2)
        self.layer_12 = nn.Conv2d(3,5,5,stride=2)
        self.layer_13 = nn.Conv2d(3,5,5,stride=2)
        self.layer_14 = nn.Conv2d(3,5,5,stride=2)
        self.layer_15 = nn.Conv2d(3,5,5,stride=2)
        self.layer_16 = nn.Conv2d(3,5,5,stride=2)
        self.layer_17 = nn.Conv2d(3,5,5,stride=2)
        self.layer_18 = nn.Conv2d(3,5,5,stride=2)
        self.layer_19 = nn.Conv2d(3,5,5,stride=2)
        self.layer_20 = nn.Conv2d(3,5,5,stride=2)
        
        self.layer_41 = nn.Linear(235200, 1)
        
        self.batch_size = batch_size
        self.length = x_train.shape[1]
        self.relu = nn.ReLU()
        self.test_on=0
        self.a=0 #1
        self.b=0 #2
        self.c=0 #3
        self.d=0 #4
        self.e=0 #5
        self.f=0 #6
        self.g=0 #7
        self.h=0 #8
        self.i=0 #9
        self.j=0 #10
        self.k=0 #11
        self.l=0 #12
        self.m=0 #13
        self.n=0 #14
        self.o=0 #15
        self.p=0 #16
        self.q=0 #17
        self.r=0 #!8
        self.s=0 #19
        self.t=0 #20
        self.u=0 #final
        
        self.a1=0 #1
        self.b1=0 #2
        self.c1=0 #3
        self.d1=0 #4
        self.e1=0 #5
        self.f1=0 #6
        self.g1=0 #7
        self.h1=0 #8
        self.i1=0 #9
        self.j1=0 #10
        self.k1=0 #11
        self.l1=0 #12
        self.m1=0 #13
        self.n1=0 #14
        self.o1=0 #15
        self.p1=0 #16
        self.q1=0 #17
        self.r1=0 #!8
        self.s1=0 #19
        self.t1=0 #20
        
        
        
        self.count_=0
        self.temp_data=0
        self.temp_result=0
        
    def forward(self, x):
        #print(x.shape)
        x_1 = self.layer_1(x[:,0,:,:,:])
        self.a = self.relu(x_1)
        x_2 = self.layer_2(x[:,1,:,:,:])
        self.b = self.relu(x_2)
        x_3 = self.layer_3(x[:,2,:,:,:])
        self.c = self.relu(x_3)
        x_4 = self.layer_4(x[:,3,:,:,:])
        self.d = self.relu(x_4)
        x_5 = self.layer_5(x[:,4,:,:,:])
        self.e = self.relu(x_5)
        x_6 = self.layer_6(x[:,5,:,:,:])
        self.f = self.relu(x_6)
        x_7 = self.layer_7(x[:,6,:,:,:])
        self.g = self.relu(x_7)
        x_8 = self.layer_8(x[:,7,:,:,:])
        self.h = self.relu(x_8)
        x_9 = self.layer_9(x[:,8,:,:,:])
        self.i = self.relu(x_9)
        x_10 = self.layer_10(x[:,9,:,:,:])
        self.j = self.relu(x_10)
        
        
        
        
        self.u = torch.cat((self.a.reshape((self.batch_size, -1)),self.b.reshape((self.batch_size, -1))),axis=1)
        self.u = torch.cat((self.u.reshape((self.batch_size, -1)),self.c.reshape((self.batch_size, -1))),axis=1)
        self.u = torch.cat((self.u.reshape((self.batch_size, -1)),self.d.reshape((self.batch_size, -1))),axis=1)
        self.u = torch.cat((self.u.reshape((self.batch_size, -1)),self.e.reshape((self.batch_size, -1))),axis=1)
        self.u = torch.cat((self.u.reshape((self.batch_size, -1)),self.f.reshape((self.batch_size, -1))),axis=1)
        self.u = torch.cat((self.u.reshape((self.batch_size, -1)),self.g.reshape((self.batch_size, -1))),axis=1)
        self.u = torch.cat((self.u.reshape((self.batch_size, -1)),self.h.reshape((self.batch_size, -1))),axis=1)
        self.u = torch.cat((self.u.reshape((self.batch_size, -1)),self.i.reshape((self.batch_size, -1))),axis=1)
        self.u = torch.cat((self.u.reshape((self.batch_size, -1)),self.j.reshape((self.batch_size, -1))),axis=1)
            
        

        x= self.layer_41(self.u)
    
        
        #x = self.relu(h_n[-1,:,:].reshape((self.batch_size,-1)))
        #x = self.layer_7(self.o[:,1:,:]).reshape((self.batch_size,-1))
        #print(x.shape)
        #x.retain_grad()
        self.count_+=1
        #print(self.count_)
        #print("종우",x.grad)
        return x

model = LRCN().cuda()
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)
#print("서",model.parameters().shape)
      
def train(dataloader_1, model, loss, optimizer):
    model.train()
    optimizer.zero_grad()
    for batch, (X, y) in enumerate(dataloader_1):
        pred = model(X)
        model.temp_data = X[0]
        #print(pred,y)
        loss_result = loss(pred, y)
        #print(model.a.grad.shape[-1])
        #print("training",pred)
        print(loss_result)
        model.temp_result=loss_result
        model.a.retain_grad()
        model.b.retain_grad()
        model.c.retain_grad()
        
        #model.d.retain_grad()
        #model.e.retain_grad()
        #model.o.retain_grad()
        #print(model.e.grad)
        loss_result.backward()
        #print(model.e.requires_grad)
        #print(model.e.grad)
        #print(model.b.grad)
        #print(model.c.grad)
        #print(model.d.grad)
        #print(model.e.grad)
        #print(loss_result.sum())
        #print(model.a.grad())
        
        
        
        #print("종우",loss_result.register_hook())
    
    #print(result_temp.shape, result_temp)
    #print(result_temp)
    #print(param.shape)
    #print(result_temp.shape)
    #print(model.)
    #print(model.count_)
    #print(model.e.grad.shape, result_temp.shape)
    if model.count_==105:
        temp_sum=0
        model.test_on=1
        #print(model.a.grad.shape)
        #width_ = model.a.grad.shape[-1]
        result_temp = torch.sum(model.a.grad.detach(),-1)
        #print(model.a.grad.detach())
        height_ = result_temp.shape[-1]
        result_temp = torch.sum(result_temp,-1)
        result_temp_ = torch.sum(result_temp,0)
        result_temp = result_temp_/5
        model.count_=0
        temp_sum=0
        #print("train",result_temp)
        #print(result_temp.shape)
        for k in range(1):  
            for j in range(result_temp.shape[0]):
                k=0
                temp_grad = model.a.grad.detach().reshape((model.batch_size, 5, model.a.grad.shape[-2],model.a.grad.shape[-1]))
                temp_sum += result_temp[j]*(temp_grad[0,j,:,:])
            result_ = temp_sum.detach().cpu().numpy()
            print("train",result_)
            min_= np.min(result_)
            result_=result_-min_
            max_ = np.max(result_)
            result_ = ((255*result_)/max_).astype(np.uint8)
            
            pil_image=Image.fromarray(result_)
            #pil_name=str(k)+str(j)+".png"
            result_=pil_image.resize((200,100))
            pil_image.save("name.png")
    
        temp_sum=0
        width_ = model.b.grad.shape[-1]
        result_temp = torch.sum(model.b.grad.detach(),-1)
        height_ = result_temp.shape[-1]
        result_temp = torch.sum(result_temp,-1)
        result_temp_ = torch.sum(result_temp,0)
        result_temp = result_temp_/5
        model.count_=0
        temp_sum=0
        for k in range(1):  
            for j in range(result_temp.shape[0]):
                k=1
                temp_grad = model.b.grad.detach().reshape((model.batch_size, 5, model.b.grad.shape[-2],model.b.grad.shape[-1]))
                temp_sum += result_temp[j]*(temp_grad[0,j,:,:])
            result_ = temp_sum.detach().cpu().numpy()
            min_= np.min(result_)
            result_=result_-min_
            max_ = np.max(result_)
            result_ = ((255*result_)/max_).astype(np.uint8)
            #for t in range(96):
                
            #    print(result_[t,:])
            pil_image=Image.fromarray(result_)
            #pil_name=str(k)+str(j)+".png"
            result_=pil_image.resize((200,100))
            pil_image.save("name_middle.png")
            
        
        
        
        #time.sleep(0.001)
        
    optimizer.step()
            
def test(dataloader_2, model, loss, optimizer):
    model.train()
    optimizer.zero_grad()
    size = len(dataloader_2.dataset)
    num_batches = len(dataloader_2)
    #model.eval()
    test_loss = 0
    #with torch.no_grad():
    for X, y in dataloader_2:
        pred = model(X)
        loss_result=loss(pred, y)
        print(loss_result)
        #print("test",pred,y)
        model.a.retain_grad()
        model.b.retain_grad()
        model.c.retain_grad()
        loss_result.backward()
    #test_loss /= num_batches
    #print(f"Test Error: \n Avg loss: {test_loss:>8f} \n")
    #print(model.test_on)
    if model.test_on==1:
        temp_sum=0
        model.test_on=0
        #print("악")
        #print(model.a.grad)
        width_ = model.a.grad.shape[-1]
        result_temp = torch.sum(model.a.grad.detach(),-1)
        height_ = result_temp.shape[-1]
        result_temp = torch.sum(result_temp,-1)
        result_temp_ = torch.sum(result_temp,0)
        result_temp = result_temp_/5
        #print("test",result_temp)
        model.count_=0
        temp_sum=0
        #print(result_temp.shape)
        for k in range(1):  
            for j in range(result_temp.shape[0]):
                k=0
                temp_grad = model.a.grad.detach().reshape((model.batch_size, 5, model.a.grad.shape[-2],model.a.grad.shape[-1]))
                temp_sum += result_temp[j]*(temp_grad[0,j,:,:])
            result_ = temp_sum.detach().cpu().numpy()
            print("test",result_)
            min_= np.min(result_)
            result_=result_-min_
            max_ = np.max(result_)
            result_ = ((255*result_)/max_).astype(np.uint8)
            
            pil_image=Image.fromarray(result_)
            #pil_name=str(k)+str(j)+".png"
            result_=pil_image.resize((200,100))
            pil_image.save("name_test.png")
    
        temp_sum=0
        width_ = model.b.grad.shape[-1]
        result_temp = torch.sum(model.b.grad.detach(),-1)
        height_ = result_temp.shape[-1]
        result_temp = torch.sum(result_temp,-1)
        result_temp_ = torch.sum(result_temp,0)
        result_temp = result_temp_/5
        model.count_=0
        temp_sum=0
        for k in range(1):  
            for j in range(result_temp.shape[0]):
                k=1
                temp_grad = model.b.grad.detach().reshape((model.batch_size, 5, model.b.grad.shape[-2],model.b.grad.shape[-1]))
                temp_sum += result_temp[j]*(temp_grad[0,j,:,:])
            result_ = temp_sum.detach().cpu().numpy()
            min_= np.min(result_)
            result_=result_-min_
            max_ = np.max(result_)
            result_ = ((255*result_)/max_).astype(np.uint8)
            #for t in range(96):
                
            #    print(result_[t,:])
            pil_image=Image.fromarray(result_)
            #pil_name=str(k)+str(j)+".png"
            result_=pil_image.resize((200,100))
            pil_image.save("name_middle_test.png")

for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(dataloader, model, loss, optimizer)
    test(dataloader_test, model, loss, optimizer)
print("Done!")
```
